# 01 — Explore Unity Catalog

Browse catalogs, schemas, and tables from your local IDE using both:
- **Spark SQL** via Databricks Connect
- **Databricks SDK** for workspace/REST operations

```python
import os
from dotenv import load_dotenv
from databricks.connect import DatabricksSession
from databricks.sdk import WorkspaceClient

load_dotenv()

spark = DatabricksSession.builder.getOrCreate()

# SDK client — uses same credential chain as Databricks Connect
w = WorkspaceClient()

# Edit these to match your Unity Catalog setup
CATALOG = os.getenv("UC_CATALOG", "main")
SCHEMA  = os.getenv("UC_SCHEMA",  "default")
```

```python
# List accessible catalogs
print("Catalogs:")
spark.sql("SHOW CATALOGS").show(truncate=False)
```

```python
# List schemas in the chosen catalog
print(f"Schemas in {CATALOG}:")
spark.sql(f"SHOW SCHEMAS IN {CATALOG}").show(truncate=False)
```

```python
# List tables in the chosen schema
print(f"Tables in {CATALOG}.{SCHEMA}:")
spark.sql(f"SHOW TABLES IN {CATALOG}.{SCHEMA}").show(truncate=False)
```

```python
# SDK example — list running clusters
print("Running clusters:")
for c in w.clusters.list():
    print(f"  {c.cluster_id}  {c.state.value:<12}  {c.cluster_name}")
```
