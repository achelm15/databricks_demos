# 00 — Verify Databricks Connect

Confirms that your local environment can reach your Databricks cluster.

**Auth options (pick one):**
1. **Databricks CLI profile** — run `databricks configure` once, then nothing else needed.
2. **Environment variables** — copy `.env.example` → `.env` and fill in values.

Databricks Connect checks in this order: env vars → `~/.databrickscfg` default profile.

```python
import os
from dotenv import load_dotenv

# Load .env if present — no-op if file doesn't exist
load_dotenv()

# Confirm key env vars are set (won't print values)
required = ["DATABRICKS_HOST", "DATABRICKS_TOKEN", "DATABRICKS_CLUSTER_ID"]
for var in required:
    status = "set" if os.getenv(var) else "NOT SET"
    print(f"{var}: {status}")
```

```python
from databricks.connect import DatabricksSession

# Picks up credentials from env vars or ~/.databrickscfg automatically
spark = DatabricksSession.builder.getOrCreate()

print(f"Spark version : {spark.version}")
print(f"Databricks host: {spark.conf.get('spark.databricks.workspaceUrl', 'n/a')}")
```

```python
# Sanity check — runs on the remote cluster, results stream back locally
df = spark.range(0, 10).withColumnRenamed("id", "value")
df.show()
```

```python
# SQL works too
spark.sql("SELECT current_user(), current_timestamp()").show(truncate=False)
```
