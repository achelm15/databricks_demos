```python
bronze_tables = list(spark.sql(f"""SHOW TABLES IN {CATALOG}.{DATABASE_B}""").select('tableName').toPandas()['tableName'].values)
silver_tables = list(spark.sql(f"""SHOW TABLES IN {CATALOG}.{DATABASE_S}""").select('tableName').toPandas()['tableName'].values)
gold_tables = list(spark.sql(f"""SHOW TABLES IN {CATALOG}.{DATABASE_G}""").select('tableName').toPandas()['tableName'].values)
```

```python
# Print details for Bronze Tables
print("Bronze Tables:\n")
for table in bronze_tables:
    # Print the table name
    print(f'Table: {table}')

    # Read the table into a DataFrame
    df = spark.read.table(f'{CATALOG}.{DATABASE_B}.{table}')

    # Print the number of rows in the table
    print(f'Number of rows: {df.count()}')

    # Print the number of unique events in the table
    print('===========================================================')
    print()

# Print details for Silver Tables
print("\nSilver Tables:\n")
for table in silver_tables:
    # Print the table name
    print(f'Table: {table}')

    # Read the table into a DataFrame
    df = spark.read.table(f'{CATALOG}.{DATABASE_S}.{table}')

    # Print the number of rows in the table
    print(f'Number of rows: {df.count()}')
    print('===========================================================')
    print()

# Print details for Gold Tables
print("\nGold Tables:\n")
for table in gold_tables:
    # Print the table name
    print(f'Table: {table}')

    # Read the table into a DataFrame
    df = spark.read.table(f'{CATALOG}.{DATABASE_G}.{table}')

    # Print the number of rows in the table
    print(f'Number of rows: {df.count()}')
    print('===========================================================')
    print()
```
