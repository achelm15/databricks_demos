# 08 — Discover Page & Databricks One Walkthrough

This notebook is a **step-by-step UI guide** for the parts of the Databricks One
experience that can't be automated via API. After running notebooks 00–07, follow
these steps to light up the full Discover + Domains + Databricks One experience.

---

## Step 1: Enable Preview Features

Both Domains and the Discover page require preview enablement.

### Account-Level (requires account admin)
1. Log in to the **Account Console** (accounts.cloud.databricks.com)
2. Go to **Settings** > **Feature enablement** > **Preview features**
3. Enable: **Domains and Discover Page**
4. Enable: **Partner-powered AI features** (required for Genie)

### Workspace-Level (requires workspace admin)
1. In the workspace, go to **Settings** (gear icon, top right)
2. Navigate to **Previews**
3. Enable: **Discover Page**

After enabling, refresh the page. "Discover" should appear in the left sidebar.

## Step 2: Create a Governed Tag

Domains are backed by a governed tag **key**. The tag key itself is the domain
identifier — any asset carrying that tag key is automatically part of the domain.

1. In the workspace, click **Catalog** in the left sidebar
2. Click the **Governed Tags** tab (at the top, next to Catalogs)
3. Click **Create governed tag**
4. Fill in:
   - **Tag key:** `nil_sponsorships`
   - **Description:** "NIL (Name, Image, Likeness) sponsorship data and analytics assets"
   - **Allowed values:** Leave empty (this tag is a simple flag, not a multi-value tag)
5. Click **Create**

> **What happens:** Notebook 05 already applied the ungoverned tag key
> `nil_sponsorships` to all silver and gold tables. Creating the governed tag
> with the same key converts those existing assignments to governed automatically.
> The tables will auto-populate in the domain you create in Step 3.

## Step 3: Create a Domain

Domains are business-aligned groupings that appear as cards on the Discover page.

1. Click **Discover** in the left sidebar (if you don't see it, re-check Step 1)
2. Click **Create domain** (top right)
3. Fill in:
   - **Domain name:** `NIL Sponsorships`
   - **Governed tag:** Select the `nil_sponsorships` tag you created in Step 2
   - **Subtitle:** "College athlete sponsorship deals, campaigns, and sponsor analytics"
   - **Description:** "A complete analytics domain for Name, Image, and Likeness (NIL) sponsorship data. Covers athlete profiles, sponsor brands, deal contracts, and marketing campaign performance across 5 conferences and 10 schools."
   - **Technical owner:** Select your user account
   - **Business owner:** Select your user account (or a colleague)
4. Click **Create**

You should now see a domain card on the Discover page.

## Step 4: Add Assets to the Domain

Tables tagged with the `nil_sponsorships` key in notebook 05 should already
appear in the domain automatically. Now add the dashboard and Genie space
(workspace assets need to be added manually).

### Add the Dashboard
1. Click into your **NIL Sponsorships** domain
2. Click **Add to Domain** (top right)
3. Switch to the **Workspace** tab
4. Search for `NIL Sponsorship Analytics` (the dashboard from notebook 06)
5. Select it and click **Add**

### Add the Genie Space
1. Still in the domain, click **Add to Domain** again
2. Switch to the **Workspace** tab
3. Search for `NIL Sponsorship Genie` (the Genie space from notebook 07)
4. Select it and click **Add**

### Verify Tables (should be auto-populated)
The gold and silver tables should already appear because they carry the
`nil_sponsorships` tag key. If any are missing:
1. Click **Add to Domain**
2. Switch to the **Catalog** tab
3. Browse to your catalog > `{schema}_gold` and select the missing tables
4. Click **Add**

## Step 5: Organize the Domain Page

Curators can create custom sections to organize assets on the domain page.

1. In the domain page, click **Edit Sections** (top right, pencil icon)
2. Create three sections:

   **Section 1: "Dashboards & Genie"**
   - Subtitle: "Interactive analytics for NIL deal insights"
   - Drag the dashboard and Genie space into this section

   **Section 2: "Core Fact Tables"**
   - Subtitle: "Deal and campaign performance data at the transaction grain"
   - Add `fact_deals` and `fact_campaign_performance`

   **Section 3: "Dimensions"**
   - Subtitle: "Reference data for athletes, sponsors, and dates"
   - Add `dim_athlete`, `dim_sponsor`, `dim_date`

3. Click **Publish** to make changes visible to all users

## Step 6: Certify Assets

Certification badges signal trust and quality to consumers. Certified assets
get priority placement on the Discover page.

### Certify the Dashboard
1. Go to the **NIL Sponsorship Analytics** dashboard
2. Click the **three-dot menu** (top right) > **Certify**
3. Confirm certification

### Certify the Genie Space
1. Navigate to the **NIL Sponsorship Genie** space
2. Click the **three-dot menu** > **Certify**
3. Confirm certification

### Certify Key Tables
1. In **Catalog Explorer**, navigate to each gold table
2. In the **About** panel (right side), click **Certify** next to the quality badge
3. Repeat for: `fact_deals`, `fact_campaign_performance`, `dim_athlete`

Certified assets will show a blue badge and appear in the "Certified" section of Discover.

## Step 7: Enable Anomaly Detection

Anomaly detection monitors data freshness and completeness, surfacing health
indicators on the Discover page.

1. In **Catalog Explorer**, navigate to your gold schema:
   `{catalog}.{schema}_gold`
2. Click the **Details** tab
3. Find **Quality monitoring** > click **Enable**
4. Toggle **Anomaly detection** on
5. Click **Save**

After a few minutes, tables will show health badges:
- **Training** — model is learning normal patterns (first few days)
- **Healthy** — metrics within expected range
- **Unhealthy** — freshness or completeness anomaly detected

> **Requires:** `MANAGE SCHEMA` or `MANAGE CATALOG` privilege on the gold schema.

## Step 8: Trigger AI-Generated Descriptions

AI descriptions auto-generate documentation from table metadata.

1. In **Catalog Explorer**, navigate to any gold table (e.g., `fact_deals`)
2. In the **About** panel, click **AI generate** (sparkle icon) next to the description
3. Review the generated description
4. Click **Accept** to save it (this runs an `ALTER TABLE ... COMMENT` under the hood)
5. For column descriptions: scroll to the column list, click **AI generate** above the header
6. Review and accept column descriptions

Repeat for key tables. These descriptions enhance search and Genie accuracy.

> **Note:** We already added manual comments in notebook 05. AI-generated
> descriptions will overwrite those — use whichever is more accurate.

## Step 9: Experience Databricks One as a Consumer

To see the full Databricks One experience:

### Option A: Toggle to Consumer View (same user)
1. Click your **profile icon** (bottom left)
2. Under **Switch experience**, select **Databricks One**
3. You should see:
   - The **NIL Sponsorships** domain card on the home page
   - The **NIL Sponsorship Analytics** dashboard in your recent/favorites
   - The **Ask** search bar at the top (connected to your Genie space)

### Option B: Use a Consumer-Only User (best for live demo)
1. In **Admin Settings** > **Users**, invite or select a test user
2. Grant only **Consumer access** entitlement (remove Workspace access and SQL access)
3. Grant the user `BROWSE` on your catalog
4. Log in as that user — they land directly in the Databricks One interface

### What to Show in the Demo
1. **Home page** — domain cards, recently shared, trending
2. **Click into the NIL Sponsorships domain** — curated sections, certified badges
3. **Open the dashboard** — interactive filters, drilldowns
4. **Use the Ask bar** — type "Which athlete has the highest total deal value?" and watch Genie answer
5. **Click into a table** — show comments, tags, lineage, health indicators
6. **Browse by domain** — show how business users navigate by function, not catalog hierarchy
