## Notebook 00: Getting Started with Databricks Connect

In this notebook, we'll kick off our Databricks Connect walkthrough by showing how to configure your environment and spin up a Spark session from your local machine, while leveraging a remote cluster behind the scenes.

```python
# Step 1: Set up the environment and connect to Databricks

# We use environment variables to securely pass credentials and cluster metadata to Databricks Connect. 
# Configuration values are pulled from a separate `config.py`.
from databricks.connect import DatabricksSession
import os
import config

os.environ["DATABRICKS_HOST"] = config.DATABRICKS_HOST
os.environ["DATABRICKS_TOKEN"] = config.DATABRICKS_TOKEN
os.environ["DATABRICKS_CLUSTER_ID"] = config.DATABRICKS_CLUSTER_ID
os.environ["DATABRICKS_WORKSPACE_ID"] = config.DATABRICKS_WORKSPACE_ID

spark = DatabricksSession.builder.getOrCreate()
```

```python
# Step 2: Sanity check — Create and display a basic DataFrame

# Let's generate a simple range of numbers to confirm everything is wired up correctly.
df = spark.range(10).withColumnRenamed("id", "value")
df.show()
```

```python
# Step 3: Run a quick distributed SQL query
spark.sql("SELECT COUNT(*) as even_count FROM range(100) WHERE id % 2 = 0").show()
```

💡 Notice that we're mixing Python and SQL here — a great example of how seamlessly Databricks Connect lets you work with both APIs in a local dev environment.
