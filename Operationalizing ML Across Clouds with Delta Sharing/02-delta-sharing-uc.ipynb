# Delta Share Creation

This notebook is responsible for creating the Delta Share used to distribute registered models across workspaces.

Key steps in this notebook:
- Create a new Delta Sharing share.
- Add registered models to the share.
- Configure permissions to allow external workspaces to access shared models.

```python
%sql
-- DELETE SHARE IF EXISTS
DROP SHARE IF EXISTS abooth_bike_sharing_model_share
```

```python
%sql
-- Create a new Delta Sharing share
CREATE SHARE abooth_bike_sharing_model_share
COMMENT "Share for production ML models.";
```

```python
%sql
-- Add a recipient. Replace the recipient name with your own. Recipients can be created via SQL or in the UI.
GRANT SELECT ON SHARE abooth_bike_sharing_model_share TO RECIPIENT `azure-field-eng-east`;
```

```python
%sql
-- Add model to the Delta Sharing share
ALTER SHARE abooth_bike_sharing_model_share
ADD MODEL alexander_booth.default.bike_sharing_uc_model;
```
