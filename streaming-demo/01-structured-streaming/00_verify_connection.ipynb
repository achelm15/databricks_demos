# 00 — Verify Connection

Confirms Databricks Connect is working against Serverless compute and that the
Unity Catalog target schema exists (or creates it).

```python
from databricks.connect import DatabricksSession
from dotenv import load_dotenv
import os

load_dotenv(dotenv_path="../.env")

spark = DatabricksSession.builder.serverless().getOrCreate()
print("Spark version:", spark.version)
```

```python
spark.sql("SELECT current_user(), current_timestamp()").show(truncate=False)
```

```python
UC_CATALOG = os.getenv("UC_CATALOG", "alexander_booth")
UC_SCHEMA  = os.getenv("UC_SCHEMA",  "streaming_demo")

spark.sql(f"CREATE SCHEMA IF NOT EXISTS {UC_CATALOG}.{UC_SCHEMA}")

print(f"Target: {UC_CATALOG}.{UC_SCHEMA}")
spark.sql(f"DESCRIBE SCHEMA {UC_CATALOG}.{UC_SCHEMA}").show(truncate=False)
```

## Reset for clean re-run (optional)

Uncomment and run the cell below to drop all tables, checkpoints, and
source files so you can re-run the full demo from scratch.

```python
# --- UNCOMMENT TO RESET ---
spark.sql(f"DROP TABLE IF EXISTS {UC_CATALOG}.{UC_SCHEMA}.slot_events_bronze")
spark.sql(f"DROP TABLE IF EXISTS {UC_CATALOG}.{UC_SCHEMA}.slot_events_silver")
spark.sql(f"DROP TABLE IF EXISTS {UC_CATALOG}.{UC_SCHEMA}.slot_events_zerobus")
spark.sql(f"DROP VOLUME IF EXISTS {UC_CATALOG}.{UC_SCHEMA}.casino_raw_events")
spark.sql(f"DROP SCHEMA IF EXISTS {UC_CATALOG}.{UC_SCHEMA} CASCADE")
spark.sql(f"CREATE SCHEMA {UC_CATALOG}.{UC_SCHEMA}")
print("Reset complete — all tables, volume, and checkpoints dropped.")
```
