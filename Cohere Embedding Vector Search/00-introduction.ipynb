# Cohere Embeddings + Databricks Vector Search
## ⚾ Scouting Report Demo

This project demonstrates how to build a semantic search pipeline using **Cohere embeddings** and **Databricks Vector Search**, applied to real-world scouting reports from **FanGraphs**.

We’ll walk through the entire lifecycle:

### 📥 1. Load Scouting Report Data
We start by pulling raw prospect data from FanGraphs, including tool grades, future value, and written scouting summaries.

### ✂️ 2. Chunk and Prepare Text
Long-form reports are normalized and chunked to fit within the 2048-character input limit of Cohere's `embed-english-v3` model.

### 🧠 3. Embed with Cohere + Delta Sync
Using Databricks Model Serving (backed by AWS Bedrock), we embed each chunk and maintain a live vector index via **Delta Sync** — no manual pipelines or orchestration required.

### 🔍 4. Search with Natural Language or Example Reports
With the vector index created, we showcase two core use cases:
- **Semantic search from a phrase** (e.g. "elite bat speed, struggles with offspeed")
- **Find similar players based on an existing report** (e.g. "show me comps for Sebastian Walcott")

This notebook sets up the environment for the rest of the workflow.
